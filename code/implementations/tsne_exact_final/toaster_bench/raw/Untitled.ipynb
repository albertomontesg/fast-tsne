```python
import os
import pandas as pd

for f in os.listdir("."):
    if f.endswith(".ipynb") or f.startswith("."):
        continue
    data = pd.DataFrame.from_csv(f, header=0, index_col=None)
    if "total" not in data:
        continue
    mat = data.as_matrix()
    new_mat = mat.copy()
    new_mat[:,-1] = mat[:,1:-1].sum(axis=1)
    new_data = pd.DataFrame(new_mat, index=None, columns=data.keys())
    new_data.to_csv(f"../{f}.csv", index=False)
    print(f)

    print(new_data.as_matrix())
```
